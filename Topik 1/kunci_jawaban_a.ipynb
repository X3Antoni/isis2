# I. Sistem Waktu-Kontinu Orde Kedua

## Nomor 1
Sebuah sistem LTI waktu-kontinu didefinisikan dalam bentuk

$$ \frac{d^2y(t)}{dt^2} + 4\frac{dy(t)}{dt}+4y(t)=x(t) $$

Sistem tersebut memiliki tanggapan frekuensi $H(j\omega)$ yang belum kita ketahui. Jika $X(j\omega)$ dan $Y(j\omega)$ eksis, maka kita bisa menggunakan sifat-sifat transformasi Fourier untuk mencari tanggapan frekuensi sistem, dimana $H(j\omega) = Y(j\omega) / X(j\omega)$.

$$ (j\omega)^2Y(j\omega)+4(j\omega)Y(j\omega)+4Y(j\omega) = X(j\omega) $$

$$ \left((j\omega)^2 + 4(j\omega) + 4 \right) Y(j\omega)=X(j\omega) $$

$$ \frac{Y(j\omega)}{X(j\omega)} = \frac{1}{(j\omega)^2+4(j\omega)+4} $$

$$ H(j\omega) = \frac{1}{(j\omega)^2+4(j\omega)+4} $$

## Nomor 2
Secara umum, tanggapan frekuensi orde dua memiliki bentuk

$$ H(j\omega) = \frac{\omega_n^2}{(j\omega)^2 + 2\zeta\omega_n(j\omega)+\omega_n^2} $$

Jika kita bandingkan bentuk umum ini dengan tanggapan frekuensi sistem sebelumnya, maka kita bisa lihat kecocokan diantara keduanya, dimana

$$ \omega_n = \sqrt{4}\;\text{rad/s} = 2\;\text{rad/s} $$

dan

$$ \zeta = \frac{4}{2\omega_n} = \frac{4}{2\cdot2\;\text{rad/s}}=1 $$

## Nomor 3
Karena $\zeta = 1$ maka sistem memililiki redaman kritis.

## Nomor 4

Untuk memplot tanggapan undak sistem, kita gunakan fungsi `step()`

```matlab
num = [0 0 1];
den = [1 4 4];
H = tf(num, den);
step(H);
```

## Nomor 5
Untuk memplot tanggapan impuls sistem, kita gunakan fungsi `impulse()`

```matlab
impulse(H);
```

## Nomor 6
Berdasarkan hasil nomor 4 dan nomor 5, kita sebenarnya tidak bisa menentukan apakah sistem tersebut mengalami redaman kritis atau redaman tinggi. Namun satu hal yang kita sudah pastikan bahwa sistem tersebut tidak mengalami redaman bawah. Hal ini tentu karena pada tanggapan undak dan impuls sistem tidak terdapat adanya osilasi.

## Nomor 7
Untuk memplot diagram Bode sistem, kita gunakan fungsi `bode()`

```matlab
bode(H)
```

## Nomor 8
Yang paling mudah adalah mencari $\omega_n$. Kita tinggal menggunakan _data cursor_, lalu tinggal menggeser-geser hingga didapatkan penurunan magnitude sebesar $-6\;\text{dB}$ (hanya untuk kasus sistem orde dua). Didapatkan $\omega_n = 2\;\text{rad/s}$.
Selanjutnya adalah mencari $\zeta$. Kita kembali perhatikan bentuk umum tanggapan frekuensi sistem orde dua.

$$ H(j\omega) = \frac{\omega_n^2}{(j\omega)^2 + 2\zeta\omega_n(j\omega)+\omega_n^2} $$

Jika $\omega = \omega_n$, maka

$$ H(j\omega_n) = \frac{\omega_n^2}{(j\omega_n)^2 + 2\zeta\omega_n(j\omega_n)+\omega_n^2} $$

$$ H(j\omega_n) = \frac{1}{j2\zeta} $$

$$ |H(j\omega_n)| = \frac{1}{2\zeta} $$

Didapatkan dari diagram Bode, $|H(j\omega_n)| = -18\;\text{dB}$. Namun agar sesuai dengan bentuk umum, kita perlu menormalisasi diagram Bode agar $|H(j\omega)_{\text{max}}| = 0\;\text{dB}$ (Perhatikan bahwa pada diagram Bode yang asli $|H(j\omega)_{\text{max}}|= -12\;\text{dB}$), sehingga $|H(j\omega_n)| =-6\;\text{dB} = -0.501$. Dengan demikian, kita dapatkan $\zeta = 1$.

## Nomor 9

Berdasarkan diagram Bode, kita dapat lihat bahwa untuk $\omega\rightarrow0$ penguatan sistem maksimum. Sebaliknya, untuk $\omega\rightarrow\infty$ penguatan sistem justru minimum dan malah mengatenuasi sinyal. Dengan demikian sistem tersebut merupakan filter _low-pass_.

# II. Tanggapan Sinusoid Sistem LTI Waktu-Kontinu

## Nomor 1

Sebuah sistem LTI waktu-kontinu didefinisikan dengan bentuk

$$ H(j\omega) = \frac{10}{j\omega+10} $$

Kita ambil bentuk magnitude-nya

$$ |H(j\omega)| = \frac{10}{\sqrt{\omega^2+100}} $$

dan bentuk fase-nya

$$ \angle H(j\omega) = -\tan^{-1}\left(\frac{\omega}{10}\right) $$

## Nomor 2

Sebuah sinyal $x(t) = \cos(t+45^\circ)$ memiliki frekuensi kerja sebesar $1\;\text{rad/s}$.

## Nomor 3
Untuk kasus $\omega = 1\;\text{rad/s}$

$$ |H(j1)| = \frac{10}{\sqrt{1^2+100}} = 0.995 $$

dan

$$ \angle H(j1) = -\tan^{-1}\left(\frac{1}{10}\right) = -5.712^\circ $$

## Nomor 4

Jika $|X(j\omega)|$ eksis, maka kita bisa mencari $Y(j\omega)$ dengan menggunakan sifat-sifat transformasi Fourier.

$$ Y(j\omega) = X(j\omega)\cdot0.995e^{-j5.712^\circ}$$

Kita transfromasi balik $Y(j\omega)$ menjadi $y(t)$

$$y(t) = 0.995\cdot x(t+39.288^\circ)$$

$$ y(t) = 0.995\cdot\cos(t + 39.288^\circ)$$

## Nomor 5

Kita buktikan apakah benar $ y(t) = 0.995\cdot\cos(t + 39.288^\circ)$?

```matlab
num = [0 10];
den = [1 10];
H = tf(num, den);
t = 0 : .2/1000 : 10;
in = cos(t + 45*pi/180);
out = lsim(H, in, t);
out = transpose(out);
plot(t, in); hold on;
plot(t, out); legend('sinyal input', 'sinyal output');
```

Kita tidak bisa begitu bedakan penguatan $1$ kali dengan $0.995$ kali. Namun kita bisa perhatikan bahwa sinyal output dan sinyal input terdapat selisih fase. Karena fase sistem negatif, maka kita dapati bahwa sinyal output pastilah _leading_ terhadap sinyal input. Hal ini sesuai dengan hasil simulasi MATLAB.

## Nomor 6

Perhatikan bahwa

$$ \angle H(j\omega) = -\tan^{-1}\left(\frac{\omega}{10}\right) $$

Sehingga sudah jelas bahwa fase-nya tidak linear. Kita akan lebih jelas memperhatikan sifat linearitas ini pada diagram Bode.

## Nomor 7

Tundaan kelompok $\tau$ didefinisikan sebagai

$$ \tau = -\frac{d}{d\omega}\left(\angle H(j\omega)\right)$$

Dengan demikian

$$ \tau = \frac{d}{d\omega}\left(\tan^{-1}\left(\frac{\omega}{10}\right)\right) = \frac{1}{\left(\frac{\omega}{10}\right)^2 + 1} = \frac{100}{\omega^2 + 100}$$

## Nomor 8

Untuk memplot diagram Bode sistem, kita gunakan fungsi `bode()`

```matlab
bode(H)
```

Perhatikan bahwa fase sistem tidak linear.

## Nomor 9

Berdasarkan diagram Bode, kita dapat lihat bahwa untuk $\omega\rightarrow0$ penguatan sistem maksimum. Sebaliknya, untuk $\omega\rightarrow\infty$ penguatan sistem justru minimum dan malah mengatenuasi sinyal. Dengan demikian sistem tersebut merupakan filter _low-pass_.

# III. Sistem Waktu-Diskret Orde Kedua

## Pendekatan Manual

Sebuah sistem LTI waktu-diskret didefinisikan dalam bentuk

$$ y[n] + y[n - 1] + \frac{1}{4} y[n - 2] = x[n]$$

Sistem tersebut memiliki tanggapan frekuensi $H(e^{j\omega})$ yang belum kita ketahui. Jika $X(e^{j\omega})$ dan $Y(e^{j\omega})$ eksis, maka kita bisa menggunakan sifat-sifat transformasi Fourier untuk mencari tanggapan frekuensi sistem, dimana $H(e^{j\omega}) = Y(e^{j\omega}) / X(e^{j\omega})$.

$$ Y(e^{j\omega}) + e^{-j\omega}Y(e^{j\omega}) + \frac{1}{4}e^{-j2\omega}Y(e^{j\omega}) = X(e^{j\omega})$$

$$ \left( 1 + e^{-j\omega} + \frac{1}{4}e^{-j2\omega} \right) Y(e^{j\omega}) = X(e^{j\omega})$$

$$ \frac{Y(e^{j\omega})}{X(e^{j\omega})} = \frac{1}{1 + e^{-j\omega} + \frac{1}{4}e^{-j2\omega}}$$

$$ H(e^{j\omega}) = \frac{1}{1 + e^{-j\omega} + \frac{1}{4}e^{-j2\omega}}$$

Bentuk umum dari sistem LTI waktu-diskret orde dua adalah

$$ H(e^{j\omega})=\frac{1}{1-2r\cos(\theta)e^{-j\omega}+r^2e^{-j2\omega}} $$

dengan demikian kita dapatkan $r = 0.5$ dan $\theta = 180^\circ$. Oleh karena $\theta = 0^\circ$ maka ada osilasi pada tanggapan sistem.

## Pendekatan MATLAB

Dengan menggunakan nilai $H(e^{j\omega})$ yang sudah dihitung sebelumnya, kita perhatikan tanggapan undak sistem

```matlab
n = 0 : 20;
num = [1 0 0];
den = [1 1 1/4];
x = ones(size(n));
y = filter(num, den, x);
stem(n,y);
```

Terlihat bahwa tanggapan undak sistem memiliki osilasi!
