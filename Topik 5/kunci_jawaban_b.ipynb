# I. Transformasi-z

## Nomor 1
Diketahui sebuah sinyal $x[n]$ didefinisikan sebagai

$$ x[n] = \left(\frac{1}{3}\right)^n \sin\left(\frac{\pi}{4}n\right)u[n] $$

Transformasi-z dari sinyal tersebut adalah

$$ X(z) = \frac{\frac{1}{3} \sin\left(\frac{\pi}{4}\right) z^{-1}}{1 - 2\cos\left(\frac{\pi}{4} \right)z^{-1} + \frac{1}{9} z^{-2}} $$

$$ X(z) = \frac{3\sin\left(\frac{\pi}{4}\right) z}{9z^2 - 2\cos\left(\frac{\pi}{4}\right) z + 1}$$

$$ X(z) = \frac{3\sqrt{2} z}{ \left(9 z^2 - 3\sqrt{2} z + 1\right) 2} $$

Hasil analisis cocok dengan MATLAB

```matlab
syms x n
x = ((1/3)^n)*sin(pi*n/4);
X = ztrans(x);
pretty(X);
```

Peta _pole-zero_

```matlab
num = [3.*sqrt(2) 0];
den = 2.*[9 -3*sqrt(2) 1];
X = tf(num, den, -1);
pzmap(X);
axis equal;
```

ROC sinyal tersebut adalah

$$ |z| > \frac{1}{3}$$

## NOMOR 2
Diketahui sebuah sinyal $x[n]$ didefinisikan sebagai

$$ x[n] = \left( -\frac{1}{3} \right)^{n} u[-n-2]$$

Transformasi-z dari sinyal tersebut adalah

$$ -\left (- \frac{1}{3} \right )^n u[-n-1] \quad \overset{z}{\longrightarrow} \quad \frac{1}{1+\frac{1}{3}z^{-1}} $$

$$ \left (- \frac{1}{3} \right )^n u[-n-1] \quad \overset{z}{\longrightarrow} \quad \frac{-1}{1+\frac{1}{3}z^{-1}} $$

$$ \left ( -\frac{1}{3} \right )^{n+1} u[-n-2] \quad \overset{z}{\longrightarrow} \quad \frac{-z}{1+\frac{1}{3}z^{-1}} $$

$$ \left ( -\frac{1}{3} \right )^{n} u[n+3] \quad \overset{z}{\longrightarrow} \quad \frac{3z}{1+\frac{1}{3}z^{-1}} $$

atau bentuk lain yaitu

$$ X(z) = \frac{3z^2}{z+\frac{1}{3}} $$

Hasil analisis ini berbeda dengan versi MATLAB

```matlab
syms x n
x = ((-1/3)^(n))*heaviside(-n-2);
X = ztrans(x);
pretty(X);
```

Peta _pole-zero_

```matlab
num = [3 0 0];
den = [0 1 1/3];
X = tf(num, den, -1);
pzmap(X);
axis equal;
```

ROC sinyal tersebut adalah

$$|z| < \frac{1}{3} $$

Diketahui sebuah sinyal $x[n]$ didefinisikan sebagai

$$ x[n] = \left( \frac{1}{2} \right)^{|n|} $$

atau cara penulisan lain adalah

$$ x[n] = \left( \frac{1}{2} \right)^{n} u[n] + \left( \frac{1}{2} \right)^{-n} u[-n-1]$$

atau

$$ x[n] = \left( \frac{1}{2} \right)^{n} u[n] - (-2^n u[-n-1])$$

Transformasi-z dari sinyal tersebut adalah

$$ X(z) = \frac{1}{1 - \frac{1}{2}z^{-1}} - \frac{1}{1 - 2z^{-1}} $$

Hasil analisis ini berbeda dengan versi MATLAB

```matlab
syms x n
x = ((1/2)^n*heaviside(n))+((1/2)^(-n)*heaviside(-n-1));
X = ztrans(x);
pretty(X);
```

Peta _pole-zero_

```matlab
num = -1;
den = [1 -2];
X1 = tf(num, den, -1);
num = 1;
den = [1 -1/2];
X2 = tf(num, den, -1);
pzmap(X1+X2);
axis equal;
```

ROC sinyal tersebut adalah

$$ \frac{1}{2} < |z| < 2 $$

$$ \delta[n-m] \quad \overset{z}{\longrightarrow} \quad z^{-m} $$

$$ x[n] = 4\delta[n+2] + 2 + 3\delta[n-1] $$

## Nomor 2
Gunakan PFE terhadap $X(z)$

```matlab
num = [1 1 0 0 0];
den = [1 4/3 -1/2 -2/3];
[R, P, K] = residuez(num, den);
R
P
K
```

Itu berarti

$$ X(z) = \frac{0.3478}{1 + 1.3333z^{-1}} + \frac{0.4183}{1 - 0.7071z^{-1}} + \frac{0.2339}{1 + 0.7071z^{-1}} $$

Sistem tersebut stabil, sehingga ROC sistem tersebut adalah

$$ |z| > 0.7071 $$

$$ |z| < 1.333 $$

maka

$$ x[n] = -0.3478(-1.333)^n u[-n-1] + 0.4183(0.7071)^n u[n] + 0.2339(-0.7071)^n u[n] $$

# III. Sistem LTI Orde-dua Waktu Diskret

## Nomor 1

Tanggapan undak untuk $r = \frac{3}{4}$ dan $\theta = 0\;\text{rad}$.

```matlab
r = 3/4;
theta = 0;

num = [1 0 0];
den = [1 -2*r*cos(theta) r^2];
H = tf(num, den, -1);
step(H);
```

```matlab
r = 3/4;
theta = pi/4;

num = [1 0 0];
den = [1 -2*r*cos(theta) r^2];
H = tf(num, den, -1);
step(H);
```

# IV. Bonus

## Sistem 1

```matlab
num = [0 1];
den = [1 8/9];
X1 = tf(num, den, -1);
pzmap(X1);
axis equal;
```

Sistem tersebut adalah kausal karena ROC-nya meliputi semua daerah lingkaran di luar lingkaran dengan jari-jari pole terluar-nya.

Sistem tersebut adalah stabil karena ROC-nya menyinggung _unit circle_.

```matlab
freqz(num, den);
```

Dari grafik yang diberikan 'kemungkinan' sistem tersebut merupakan berupa filter _high-pass_.
