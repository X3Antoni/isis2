# I. Jaringan Crossover Pasif


## Nomor 1.1
Gunakan _voltage-divider_ untuk mendapatkan fungsi alih sistem I

$$ H(s) = \frac{R\parallel \frac{1}{sC}}{\left( R \parallel \frac{1}{sC} \right) + sL} $$

$$ H(s) = \frac{\frac{R}{sRC + 1}}{\frac{R}{sRC + 1} + sL} $$

$$ H(s) = \frac{\frac{R}{sRC + 1}}{\frac{R}{sRC + 1} + \frac{s^2RLC + sL}{sRC + 1}} $$

$$ H(s) = \frac{R}{s^2RLC + sL + R}$$

$$ H(s) = \frac{10}{s^2\left( 2.533\times10^{-13} \right) + s \left( 2.508\times10^{-6} \right) + 10} $$

## Nomor 1.2
Ada 2 buah pole pada sistem tersebut, masing-masing terletak di $(-4.429 \pm j4.429)\times10^6$

Ada 2 buah zero pada sistem tersebut, keduanya terletak di $\infty$.

```matlab
R = 10;
L = 2.2508*10^-6;
C = 1.1254*10^-8;
num = R;
den = [R*L*C L R];
H1 = tf(num, den);
zero = roots(num)
pole = roots(den)
```

## Nomor 1.3
Peta _pole-zero_

```matlab
pzmap(H1)
```

## Nomor 1.4
Ada 2 buah skenario ROC yang mungkin untuk sistem tersebut.

Skenario yang kedua adalah untuk kasus sistem yang tidak-stabil dan anti-kausal. Jika sistem tersebut adalah sistem yang demikian, maka ROC sistem tersebut adalah $\Re(s) < -4.4429\times10^6$.

Berdasarkan plot Bode sistem I, terlihat bahwa sistem tersebut merupakan filter _high-pass_.

# II. Lokasi Pole Sistem Orde-Dua

## Nomor 2.1

```matlab
zeta = linspace(0, 4, 20);
wn = 1;

den = zeros(1, length(zeta));
for i = 1 : length(zeta)
    num = wn^2;
    den(1, i) = 1;
    den(2, i) = 2*zeta(1,i)*wn;
    den(3, i) = wn^2;
end

for i = 1 : length(zeta)
    H = tf(num, transpose(den(:,i)));
    pzmap(H);
    hold on;
    axis equal;
    axis([-4 0 -2 2]);
    % k = waitforbuttonpress; % optional saja u/ keperluan animasi
end
```

Seperti yang sudah pernah kita pelajari sebelumnya, $\zeta$ mempengaruhi karakteristik redaman sistem.

Untuk $\zeta = 0$, kita perhatikan pole-nya murni imajiner! Itu berarti sistem tersebut menghasilkan tanggapan yang bersifat osilasi tanpa ada redaman.

Untuk $0 < \zeta < 1 $, kita perhatikan bahwa pole-nya mulai semakin ke kanan, namun sekarang tidak lagi murni imajiner. Itu berarti sistem tersebut tetap bersifat osilasi namun dengan adanya sedikit redaman.

Untuk $\zeta = 1$, kita perhatikan bahwa pole-nya terletak pada titik yang sama! (anda bisa modifikasi sedikit array zeta tersebut sehingga terdapat $\zeta = 1$). Itu berarti, sistem mulai tidak bersifat osilasi. Kita katakan redaman pada saat ini merupakan redaman kritis.

Untuk $\zeta > 1$, kita perhatikan bahwa pole-nya murni real! Itu berarti sistem tersebut tidak lagi memiliki tanggapan yang bersifat osilasi, namun sebaliknya memiliki redaman yang cukup besar.

## Nomor 2.2

```matlab
zeta = 1/4;
wn = linspace(0, 4, 20);

num = zeros(1, length(wn));
den = zeros(3, length(wn));
for i = 1 : length(wn)
    num(1, i) = wn(1,i)^2;
    den(1, i) = 1;
    den(2, i) = 2*zeta*wn(1,i);
    den(3, i) = wn(1,i)^2;
end

for i = 1 : length(wn)
    H = tf(num(1,i), transpose(den(:,i)));
    pzmap(H);
    hold on;
    axis equal;
    axis([-4 0 -2 2]);
    % k = waitforbuttonpress; % optional saja u/ keperluan animasi
end
```

Seperti yang kita pelajari sebelumnya bahwa $\omega_n$ sebenarnya merupakan frekuensi natural sistem. Semakin tinggi $\omega_n$, maka semakin besar pula magnitude pole-nya. Kasus khusus untuk $\omega_n = 0$, kita perhatikan bahwa pole sistem pas terletak di titik _origin_, _which basically mean_ sistem tersebut adalah filter _all-reject_ (menolak semua jenis sinyal input, $H(s) = 0$).

## Nomor 2.3

```matlab
zeta = linspace(-2, 2, 20);
wn = 1;

den = zeros(1, length(zeta));
for i = 1 : length(zeta)
    num = wn^2;
    den(1, i) = 1;
    den(2, i) = 2*zeta(1,i)*wn;
    den(3, i) = wn^2;
end

for i = 1 : length(zeta)
    H = tf(num, transpose(den(:,i)));
    pzmap(H);
    hold on;
    axis equal;
    axis([-2 2 -2 2]);
    % k = waitforbuttonpress; % optional saja u/ keperluan animasi
end
```

Seperti yang sudah pernah kita pelajari sebelumnya, $\zeta$ mempengaruhi karakteristik redaman sistem.

Untuk $\zeta = 0$, kita perhatikan pole-nya murni imajiner! Itu berarti sistem tersebut menghasilkan tanggapan yang bersifat osilasi tanpa ada redaman.

Untuk $0 > \zeta > -1 $, kita perhatikan bahwa pole-nya mulai semakin ke kanan, namun sekarang tidak lagi murni imajiner. Itu berarti sistem tersebut tetap bersifat osilasi namun dengan adanya sedikit redaman.

Untuk $\zeta = -1$, kita perhatikan bahwa pole-nya terletak pada titik yang sama! (anda bisa modifikasi sedikit array zeta tersebut sehingga terdapat $\zeta = 1$). Itu berarti, sistem mulai tidak bersifat osilasi. Kita katakan redaman pada saat ini merupakan redaman kritis.

Untuk $\zeta < -1$, kita perhatikan bahwa pole-nya murni real! Itu berarti sistem tersebut tidak lagi memiliki tanggapan yang bersifat osilasi, namun sebaliknya memiliki redaman yang cukup besar.

Tentu perbedaannya terletak di sifat sistem itu sendiri! Untuk kasus nomor 1, kita asumsikan bahwa sistem kausal, oleh karenanya untuk sembarang $\zeta$ bernilai positif maka sistem akan mengalami ketidakstabilan. Untuk kasus nomor 2, kita asumsikan bahwa sistem anti-kausal, oleh karenanya untuk sembarang $\zeta$ bernilai negatif maka sistem akan mengalami ketidakstabilan.
